```python
import os 
import pandas as pd


from pysentimiento import SentimentAnalyzer
from pysentimiento import EmotionAnalyzer
from pysentimiento.preprocessing import preprocess_tweet
```

```python
df_non = pd.read_csv('df_non.csv')
df_music = pd.read_csv('df_music.csv')

#trying_df = df_non.head(10).copy(deep=True) #until i run on ucloud i'll just do it on a segment of tweets

```

```python

#importing sentiment function from pipeline
analyzer = SentimentAnalyzer(lang="en")
emotion_analyzer = EmotionAnalyzer(lang="en")
```

```python

analyzer.predict("Fears Strictly Come Dancing stars will be 'locked out' of UK during coronavirus")

emotion_analyzer.predict("Fears Strictly Come Dancing stars will be 'locked out' of UK during coronavirus")


```

```python
# first non music related

# new columns 
preprocessed = []
sentiment = []
sentiment_probas = []
emotion = []
emotion_probas = []

for index, row in df_non.iterrows(): 
    preprocessed_text = preprocess_tweet(row["texts"])
    preprocessed.append(preprocessed_text)
    
    sentiment_prediction = analyzer.predict(preprocessed_text)
    sentiment.append(sentiment_prediction.output)
    sentiment_probas.append(sentiment_prediction.probas)
    
    emotion_prediction = emotion_analyzer.predict(preprocessed_text)
    emotion.append(emotion_prediction.output)
    emotion_probas.append(emotion_prediction.probas)
    
# add columns to df 
df_non["preprocessed_text"] = preprocessed
df_non["sentiment_classification"] = sentiment
df_non["sentiment_scores"] = sentiment_probas
df_non["emotion_classification"] = emotion 
df_non["emotion_scores"] = emotion_probas
```

```python
#print(trying_df)

df_non.head(20)
print(df_non["emotion_classification"].value_counts())
print(df_non["sentiment_classification"].value_counts())
```

```python
# then music related

# new columns 
preprocessed = []
sentiment = []
sentiment_probas = []
emotion = []
emotion_probas = []

for index, row in df_music.iterrows(): 
    preprocessed_text = preprocess_tweet(row["texts"])
    preprocessed.append(preprocessed_text)
    
    sentiment_prediction = analyzer.predict(preprocessed_text)
    sentiment.append(sentiment_prediction.output)
    sentiment_probas.append(sentiment_prediction.probas)
    
    emotion_prediction = emotion_analyzer.predict(preprocessed_text)
    emotion.append(emotion_prediction.output)
    emotion_probas.append(emotion_prediction.probas)
    
# add columns to df 
df_music["preprocessed_text"] = preprocessed
df_music["sentiment_classification"] = sentiment
df_music["sentiment_scores"] = sentiment_probas
df_music["emotion_classification"] = emotion 
df_music["emotion_scores"] = emotion_probas
```

```python
#print(trying_df)

df_music.head(20)
print(df_music["emotion_classification"].value_counts())
print(df_music["sentiment_classification"].value_counts())
```
